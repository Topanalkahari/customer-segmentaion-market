```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
```

```python
label_encoder = LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender']) # Apply label encoding to 'gender' column
df['preferred_category'] = label_encoder.fit_transform(df['preferred_category']) # Apply label encoding to 'preferred_category' column
```

```python
# Inisialisasi OneHotEncoder
one_hot = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

categorical_features_onehot = ['gender', 'preferred_category']
# Select the actual data from the DataFrame
categorical_features_data = df[categorical_features_onehot]
# Melakukan transformasi OneHotEncoding
categorical_features_encode =  one_hot.fit_transform(categorical_features_data)
# Mengubah hasil encoding menjadi dataframe
categorical_encoded_df = pd.DataFrame(categorical_features_encode, columns=one_hot.get_feature_names_out(categorical_features_data.columns))

# Menggabungkan dataframe hasil encoding OneHot dengan kolom 'trip_type' yang telah diubah
df_modified = df.drop(['gender', 'preferred_category'], axis=1)
df_modified = pd.concat([df_modified, categorical_encoded_df], axis=1)
df = df_modified
```

```python
df = df.drop(['id'], axis=1)
```
